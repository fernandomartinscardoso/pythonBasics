## Comprehensive List and File Manipulation

### Comprehensive list

Let's start with comprehensive list, a method that allows us to create sequences based on other lists in a very easy and sinthetic way.

```python
numbers = [0, 1, 2, 3, 4]
squares = []
for i in numbers:
    squares.append(i**2)
print(squares)
```

```python
numbers = [0, 1, 2, 3, 4]
squares = [number**2 for number in numbers]
print(squares)
```

The code with comprehensive list is more concise and easier to code. :-)

### File Manipulation

Reading and right files, also using the method `with ... as`

```python
import random

file = open("numbers.txt","w")
file.write("") # Cleaning previous values
file.close()
file = open("numbers.txt","a")
for j in range(100):
    rand_int = random.randint(0, 1000)
    file.write(str(rand_int))
    file.write("\n")
file.close()
```

```python
numbers_list = []
with open("numbers.txt","r") as file:
    for k in range(100):
        number = int(file.readline().strip("\n"))
        numbers_list.append(number)
print(numbers_list)
```
